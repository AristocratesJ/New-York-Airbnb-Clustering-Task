```python
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
df = pd.read_csv("./new_york_listings_2024.csv")
df
```

```python
df_model, df_val = train_test_split(df, test_size=0.2, random_state=10)
```

```python
df_model = df_model.reset_index(drop=True)
df_model
```

```python
df_val = df_val.reset_index(drop=True)
df_val
```

```python
df_model.to_csv("./to_model/new_york_listings_2024_to_model.csv", index=False)
df_val.to_csv("./to_val/new_york_listings_2024_to_val.csv", index=False)
```
